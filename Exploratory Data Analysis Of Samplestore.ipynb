# Exploratory Data analysis And Visualisation of Online Sample Store 

In this Analysis i made use of a sample store dataset gotten from kaggle and performed some exploratory analysis to gain insightson some Kpi's, which would be very useful in real world scenarios. I was able to gain the total number of all coumtries that bought from the store.
The top 5 highest and lowest selling products, and also the top 5 highest prchasing countries. All my steps are shown below and explained in full detail

```python
## Import all neccessary libraries to be used for our analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## read in our dataset which is an excel workbook 
df=pd.read_excel('Online Retail Data Set (1).xlsx')
```

```python
## Check our dataset
df
```

```python
## Here i created a new column whcih is a multiplication of the quantity and unitprice to derive a totalsales 
df['TotalSales']=df['Quantity'] * df['UnitPrice']
```

```python
## Checked the datset and all its components 
df.head
```

```python
## Removed a column that wouldnt be useful in our analysis 
df = df.drop('StockCode',axis=1)
```

```python
df
```

```python
## Checked for null data
df.isnull()
```

```python
## Using unique() to check the countries in the dataset 
unique_countries =df['Country'].unique()
```

```python
print(unique_countries)
```

```python
## used nunique()which gave us the total amount of countries in our dataset
unique_country_count = df['Country'].nunique()
```

```python
print(unique_country_count)
```

```python
## Used value_counts() to check the total amount of each countries in the dataset 
unique_country_value_count = df['Country'].value_counts()
```

```python
print(unique_country_value_count)
```

```python
## Also used value_count to check for the goods and its total amount 
description_count = df['Description'].value_counts()
```

```python
print(description_count.head(5))
```

```python
print(description_count.tail(5))
```

```python
## Used sum() to calculate total sales 
Total_orders = df['Quantity'].sum()
```

```python
print(Total_orders)
```

```python
## converted the InvoiceDate column to a datetime type using the pd.to.datetime() method. 
df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'])
```

```python
## I made use of the 'dt' accesor to extract the date and time components 
## from the column and create two new columns named 'date and 'time'

df['date']= df['InvoiceDate'].dt.date
df['time']= df['InvoiceDate'].dt.time
```

```python
df
```

```python
## Made use od drop() to remove the invoicedate column 
## as it will not be needed since we have derived the "date",'time',and 'year'

df=df.drop('InvoiceDate', axis=1)
```

```python
df
```

```python
## Here i made use of groupby() to group the countries and also added the total sales of all the different countries 
grouped =df.groupby('Country')['TotalSales'].sum()
```

```python
## Made use of nlargest() to derive the top 5 highest selling countries 
top_countries =grouped.nlargest(5)
```

```python
## Here i made use of matplotlib library to plot a barchat visualizing the Top 5 countries per Total sales 
## and as we can see the uited kingdom has the highest sales 
plt.bar(top_countries.index,top_countries.values, color='red')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total sales ')
plt.title('Top 5 countries per Total Sales ')
```

```python
## I made use of the 'pd.to_datetime' to derive the 'years' from the date column
df['date'] =pd.to_datetime(df['date'], format ='%Y-%m-%d')
df['year'] =df['date'].dt.year
```
